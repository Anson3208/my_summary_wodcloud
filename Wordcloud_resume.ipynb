```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import re
```

```python
# import resume
resume_raw = open("resume.txt", encoding="utf-8").read()
```

```python
# Transform all the words to lowercase

lowercase_resume = resume_raw.lower()
re_pattern = r'[A-Za-z0-9]+' 
raw_tokens = re.findall(re_pattern,lowercase_resume)

stopwords = set(STOPWORDS)
clean_tokens = []
```

```python
for token in raw_tokens:
    if token in raw_tokens:
        if token not in stopwords:
            clean_tokens.append(token)
```

```python
# Generate the text
resume = ' '.join(clean_tokens)

from PIL import Image
mask = np.array(Image.open('img/cloud_02.png'))

# Create a WordCloud object with transparent background and RGBA mode
cloud = WordCloud(background_color=(255, 255, 255, 0), width=1600, height=800, max_words=1000, mask=mask, margin=1)

# Generate a word cloud image
cloud_img = cloud.generate(resume)

# Visualize the word cloud image using matplotlib
plt.figure(figsize=(16, 8), dpi=200)
plt.imshow(cloud_img, interpolation='bilinear')
plt.axis('off')
plt.show()
```
